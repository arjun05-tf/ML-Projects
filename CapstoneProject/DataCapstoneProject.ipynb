```python
#importing required libraries 
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("Countries.csv")
```

```python
df.head()
```

```python
#Which country has the highest population?
df[df["population"] == df["population"].max()]["country"]
```

```python
#What is the capital of the country with the highest population? 
df[df["population"] == df["population"].max()][["country", "capital_city"]]
```

```python
#Which country has the least population? 
df[df["population"] == df["population"].min()]["country"]
```

```python
#What is the capital of the country with the least population? 
df[df["population"] == df["population"].min()][["country", "capital_city"]]
```

```python
#What are the top 5 countries with highest democratic scores?
df.sort_values("democracy_score", ascending = False, inplace = True)
df["country"].head()
```

```python
#How many total regions are there?
df["region"].value_counts().count()
```

```python
#How many countries lie in the Eastern European region?
df["region"].value_counts()["Eastern Europe"]
```

```python
#Who is the political leader of the 2nd most populated country?
df[df["population"] == df["population"].nlargest(2).iloc[1]]["political_leader"]
```

```python
#How many countries are there whose political leaders are unknown?
df[df["political_leader"].isna()]["country"].count()
```

```python
#How many countries have republic in their fullname?
country_name = df["country_long"]
```

```python
count = df["country_long"].str.lower().str.contains("republic").sum()
print(count)
```

```python
#Which country in Africa region has the highest population?
africa_df = df[df["continent"] == "Africa"]
africa_df[africa_df["population"] == africa_df["population"].max()]["country"]
```
